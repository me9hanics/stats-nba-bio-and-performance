# Fetching the data of players from NBA.com, through the nba_api package

To access the nba_api package, it needs to be installed with pip. Therefore I don't recommend using conda for loading the environment, as you could create conflicts with pip and conda package managers at the same time. I recommend using pyenv (with virtualenv), then after creating a new virtual environment, install the package with pip:

```powershell

pip install nba_api

```

I used Windows, on it it's a bit more complicated. Install pyenv-win, then find something to create pyenv virtual environments with (I used https://github.com/pyenv-win/pyenv-win-venv), create and switch to a new virtual environment, then install pip and the package with pip. In this case, you have to run:

```powershell
pyenv exec pip install nba_api
```

to install the package in the virtual environment.

````python
import pandas as pd
import numpy as np
````

````python
from nba_api.stats.static import players

# get_players returns a list of dictionaries, each representing a player.
nba_players = players.get_players()
print("Number of players fetched: {}".format(len(nba_players)))
nba_players[0]
````

````python
ids =[player['id'] for player in nba_players]
len(ids)
````

````python
from nba_api.stats.endpoints import playercareerstats as career
id = 76001
career_stats = career.PlayerCareerStats(player_id=id, timeout=2)
careers = career_stats.get_data_frames()[0]
````

````python
id_len = len(ids)
ids_half = ids[:id_len//2]
ids_half2 = ids[id_len//2:]
ids_quarter1 = ids[:id_len//4]
ids_quarter2 = ids[id_len//4:id_len//2]
ids_quarter3 = ids[id_len//2:3*id_len//4]
ids_quarter4 = ids[3*id_len//4:]

wrong = []
````

````python
#NOTE: Since the commands below were ran, I changed some parts of the function: made careers and wrong not global vars but arguments, and removed first warning.

def get_careers(ids, careers, wrong):
    for id in ids:
        try:
            career_stats = career.PlayerCareerStats(player_id=id, timeout=2)
            careers.append(career_stats.get_data_frames()[0])
        except:
            #NOTE: Here one should add a time.sleep(1)
            try:
                career_stats = career.PlayerCareerStats(player_id=id, timeout=2)
                careers.append(career_stats.get_data_frames()[0])
            except:
                wrong.append(id)
    return wrong
````

````python
bio_stats = bio.LeagueDashPlayerBioStats(season="2022-23", timeout = 2)  #LeagueID, PerMode, Season, SeasonType
bios = bio_stats.get_data_frames()[0]
bios["SEASON_ID"] = ["2022-23"]*len(bios)
bios
````

````python
bio_stats = bio.LeagueDashPlayerBioStats(season="2022-23", timeout = 2)  #LeagueID, PerMode, Season, SeasonType
bios = bio_stats.get_data_frames()[0]
#bios["SEASON_ID"] = ["2022-23"]*len(bios)
bios
````

````python
import time
seasons = ["2021-22","2020-21",'2019-20', '2018-19', '2017-18', '2016-17', '2015-16', '2014-15', '2013-14','2012-13', '2011-12', '2010-11', '2009-10', '2008-09','2007-08', '2006-07', '2005-06', '2004-05', '2003-04',]
wrongs = []
for season in seasons:
    try:
        bio_stats = bio.LeagueDashPlayerBioStats(season=season, timeout = 4)
        #bio_stats["SEASON_ID"] = [season]*len(bio_stats)
        bios = pd.concat([bios, bio_stats.get_data_frames()[0] ], axis=0, ignore_index=True)
    except:
        time.sleep(1)
        try:
            bio_stats = bio.LeagueDashPlayerBioStats(season=season, timeout = 4)
            #bio_stats["SEASON_ID"] = [season]*len(bio_stats)
            bios = pd.concat([bios, bio_stats.get_data_frames()[0] ], axis=0, ignore_index=True)
        except:
            wrongs.append(season)
````
